```python
# Create 4 arrays to populate
# Will later create a dataframe
tracks = []
lyrics = []
views = []
popularity_index = []
count = 1 # Will serve as Popularity Index 
# Search Genius Database for songs (Sorted by popularity as default)
songs = (genius.search_artist(artist, max_songs=50,get_full_info=False)).songs
for song in songs:
    title = song.title.replace('\'','')
    title = title.replace('’','')
    #title = title.split(' (')[0]  /// Just in case a title goes by another name
    tracks.append(title)
    # Cleans lyrics since they have paragraphs as \n
    clean = song.lyrics.replace('\n', ' ')
    clean2 = clean.replace('\'', '')
    lyrics.append(clean2) 
    popularity_index.append(count)
    # IF stats.pageviews not found 
    #try:
    views.append(song.stats.pageviews)
    #except:
    #    views.append(0)
    #    break
    count += 1
# Creation of DataFrame
artist_data = pd.DataFrame(index=tracks)
artist_data['Lyrics']=lyrics
artist_data['PopIndex'] = popularity_index
artist_data['PageViews'] = views
## DF created with Name - Lyrics - Index (Pop. Num)
## Removed ' and '’' characters
```

```python
### CHANGE DF VARIABLES

# Main DF - shallow copy of dataframe to make changes
main = artist_data

# New arrays to be added onto DataFrame for each song
popularity = []
dance = []
energy = []
loudness = []
valence = []
tempo = []
instru = []
speech = []


missing = 0
for song in main.index:
    i = 0
    #Search Spotify for song + artist
    track = spot.search(q=str(song + ' ' + artist))
    try:
        song_artist = track['tracks']['items'][0]['album']['artists'][0]['name']
        # Loop to find song by correct artist
        while song_artist != artist:
            #DEBUGGING
            #print(song_artist)
            i += 1
            song_artist = track['tracks']['items'][i]['album']['artists'][0]['name']
            if i > 4:
                # Give up - song is not in there
                break
        song_id = track['tracks']['items'][0]['id']
    # If song + artist not found, go to ELSE statement and add all zeroes for later filtering
    except:
        song_artist = ''
        pass
    if song_artist == artist:
        track = spot.search(q=str(song + ' ' + artist))
        dance.append(spot.audio_features(song_id)[0]['danceability'])
        popularity.append(track['tracks']['items'][0]['popularity'])
        energy.append(spot.audio_features(song_id)[0]['energy'])
        loudness.append(spot.audio_features(song_id)[0]['loudness'])
        valence.append(spot.audio_features(song_id)[0]['valence'])
        tempo.append(spot.audio_features(song_id)[0]['tempo'])
        instru.append(spot.audio_features(song_id)[0]['instrumentalness'])
        speech.append(spot.audio_features(song_id)[0]['speechiness'])
    # Song is not found on Spotify; likely Soundcloud only and not on streaming services
    # Handle by adding 0 to song characteristics to filter out later
    else: 
        popularity.append(0)
        dance.append(0)
        energy.append(0)
        loudness.append(0)
        valence.append(0)
        tempo.append(0)
        instru.append(0)
        speech.append(0)
        # Prints song name if it is not here
        missing += 1
        print(str(missing) + ' Missing: ' + song)
# Add arrays of song characteristics into main DataFrame
main['popularity'] = popularity
main['dance'] = dance
main['energy'] = energy
main['loudness'] = loudness
main['valence'] = valence
main['tempo'] = tempo
main['instru'] = instru
main['speech'] = speech
# Length that has the proper characteristics
trimmed_length = len(main.index)-missing
print(trimmed_length)
```
